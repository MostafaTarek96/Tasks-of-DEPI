```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('diamonds.csv')
df
```

```python
print(df.isnull().sum())
```

```python
df['cut'] = df['cut'].str.lower()
df = df[df['cut'].notna()]
df=df.drop(columns=['Unnamed: 0'])
df
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='carat', y='price', data=df)
plt.title('relation between carat and price')
plt.xlabel('carat')
plt.ylabel('price')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x='cut', y='price', data=df)
plt.title('relation between cut and price')
plt.xlabel('cut')
plt.ylabel('price')
plt.show()
```

```python
sns.countplot(x='cut', data=df,hue='color')
plt.title('count of cut and colors')
plt.xlabel('cut')
plt.ylabel('count')
plt.show()
```

```python
sns.barplot(x='clarity', y='price', data=df)
plt.title('relation between clarity and price')
plt.xlabel('clarity')
plt.ylabel('price')
plt.show()
```
