```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns

titanic=sns.load_dataset('titanic')
titanic
```

```python
print(titanic.isnull().sum())
```

```python
titanic.drop(columns=['deck'], inplace=True)
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.title('Relation between Fare, Survival, and Sex')
plt.show()
```

```python
sns.countplot(data=titanic,x='survived',hue='class')
plt.title('Survival Count by Class')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='sex', y='age', data=titanic)
plt.title('Age Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
sns.countplot(data=titanic,x='class',hue='survived')
plt.title('Class Distribution by Sex')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(data=titanic,x='survived',y='age')
plt.title('Age Distribution by Survival and Sex')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
x=titanic.groupby(['sex','class'])['survived'].sum()
y=[f"{sex}-{classs}"for sex,classs in x.index]
z=x.values
plt.pie(z,labels=y,autopct='%1.1f%%')
plt.title('Survived Distribution by Sex')
```
